# Logistic Regression

## Importing the libraries

```
import pandas as pd
```

## Importing the dataset

```
dataset = pd.read_csv('breast_cancer.csv')
dataset_n = pd.read_csv('wdbc.csv')
#dataset_n.info()
print(dataset_n.iloc[:,1].unique())
X_n = dataset_n.iloc[:, 2:].values
y_n = dataset_n.iloc[:, 1].values


#print(dataset.isna().sum())
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```
print(y_n)
```

```
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_n = le.fit_transform(y_n)
```

```
print(y_n)
```

## Splitting the dataset into the Training set and Test set

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
X_n_train, X_n_test, y_n_train, y_n_test = train_test_split(X_n, y_n, test_size = 0.2, random_state = 0)
```

## Feature Scalling

```
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_n_train = sc.fit_transform(X_n_train)
X_n_test = sc.transform(X_n_test)
```

## Training the Logistic Regression model on the Training set

```
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier_n = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
classifier_n.fit(X_n_train, y_n_train)
```

## Predicting the Test set results

```
y_pred = classifier.predict(X_test)
Y_n_pred = classifier_n.predict(X_n_test)
```

## Making the Confusion Matrix

```
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm_n = confusion_matrix(y_n_test, Y_n_pred)
#print(cm)
print(cm_n)
```

## Computing the accuracy with k-Fold Cross Validation

```
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = classifier, X = X_train, y = y_train, cv = 10)
accuracies_n = cross_val_score(estimator = classifier_n, X = X_n_train, y = y_n_train, cv = 5)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
print("Accuracy: {:.2f} %".format(accuracies_n.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies_n.std()*100))
```
