# Prepare the data for model training and split it into training and testing

```python
# Let's see how many customers could potentially default on their credit card payment!
uci_default_df        = uci_df[uci_df['default.payment.next.month'] == 1]
uci_nodefault_df      = uci_df[uci_df['default.payment.next.month'] == 0]
```

```python
from sklearn.preprocessing import OneHotEncoder

cat = uci_df[['SEX', 'EDUCATION', 'MARRIAGE']]
print(cat.info())
encoder = OneHotEncoder(sparse=False)
cat = encoder.fit_transform(cat)
cat.shape
cat = pd.DataFrame(cat)
cat.info()
```

```python
num = uci_df.drop(columns=['SEX', 'EDUCATION', 'MARRIAGE', 'ID', 'default.payment.next.month'], axis = 1)
num.info()
```

```python
X = pd.concat([num, cat], axis=1)
X.columns = X.columns.astype(str)
print(X.head())
y = uci_df['default.payment.next.month']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
X_train.shape, X_test.shape
```

```python
# Because w have aultliers using StandardScaler can impact the ouput
# because of mean valuse usage. So we need to Normalization

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
X_train.shape, X_test.shape
```

```python
X_train = pd.DataFrame(X_train)
X_test = pd.DataFrame(X_test)
# X_train.columns = X_train.columns.astype(str)
# X_test.columns = X_test.columns.astype(str)
# X_train.describe()
```

```python
# Since we have Multicollinearity let try to reduce dimensions

from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
n_features = 46
n_classes = len(np.unique(y_train))
# print(min(n_features, n_classes-1))
lda = LDA(n_components = 1)
X_train_lda = lda.fit_transform(X_train, y_train) # the only difference in implementation with Principal Component analysis

X_test_lda = lda.transform(X_test)
X_train_lda.shape, X_test_lda.shape
```

# Train and evaluate an XG-Boost classifier model

```python
from xgboost import XGBClassifier

xgb_model = XGBClassifier(learning_rate = 0.1, max_depth = 40, n_estimators = 1000) #objective="binary:logistic", max_depth=10, n_estimators = 300
xgb_model.fit(X_train_lda, y_train)

# xgb_model_lda = XGBClassifier()
# xgb_model_lda.fit(X_train_lda, y_train)
```

```python
score = xgb_model.score(X_test_lda, y_test)
# score_lda = xgb_model_lda.score(X_test_lda, y_test)
print(score)
# print(score_lda)
```

```python
# Plot the feature importance

feat_scores= pd.DataFrame({"Fraction of variables affected" : xgb_model.feature_importances_},index = X.columns)
feat_scores= feat_scores.sort_values(by = "Fraction of variables affected")
feat_scores.plot(kind = "barh", figsize = (20, 15))
sns.despine()
```

```python
xgb_pred = xgb_model.predict(X_test_lda)
# xgb_model_lda.predict(X_test_lda)


print(confusion_matrix(y_test, xgb_model.predict(X_test_lda)))
print(classification_report(y_test, xgb_pred))
# confusion_matrix(y_test, xgb_model_lda.predict(X_test_lda))
```

```python
# ROC curve
from sklearn.metrics import roc_curve

fpr1, tpr1, thresh1 = roc_curve(y_test, xgb_model.predict_proba(X_test)[:, 1], pos_label = 1)
# fpr1_lda, tpr1_lda, thresh1_lda = roc_curve(y_test, xgb_model_lda.predict_proba(X_test_lda)[:, 1], pos_label = 1)
# print(model_svm.predict_proba(X_test))
# print(svc_model.predict_proba(X_test)[:, 1])

fpr2, tpr2, thresh2 = roc_curve(y_test, model_svm.predict_proba(X_test)[:, 1], pos_label = 1)
fpr3, tpr3, thresh3 = roc_curve(y_test, gnb_model.predict_proba(X_test)[:, 1], pos_label = 1)
fpr4, tpr4, thresh4 = roc_curve(y_test, log_model.predict_proba(X_test)[:, 1], pos_label = 1)
fpr5, tpr5, thresh5 = roc_curve(y_test, rf_model.predict_proba(X_test)[:, 1], pos_label = 1)
fpr6, tpr6, thresh6 = roc_curve(y_test, knn_model.predict_proba(X_test)[:, 1], pos_label = 1)

plt.plot(fpr1, tpr1, linestyle = "--", color = "orange", label = "XgBoost")
plt.plot(fpr2, tpr2, linestyle = "--", color = "red", label = "SVM")
plt.plot(fpr3, tpr3, linestyle = "--", color = "green", label = "Naive bayes")
plt.plot(fpr4, tpr4, linestyle = "--", color = "blue", label = "logistic")
plt.plot(fpr5, tpr5, linestyle = "--", color = "black", label = "Random forest")
plt.plot(fpr6, tpr6, linestyle = "--", color = "yellow", label = "K-Nearest Neighbors")

plt.title('Receiver Operator Characteristics (ROC)')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive rate')

plt.legend(loc = 'best')
plt.savefig('ROC', dpi = 300)
plt.show()
```
